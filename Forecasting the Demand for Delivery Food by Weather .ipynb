```python
a =jajang.pivot_table('통화건수', index=['기준일','시군구','성별','요일','연령대'], aggfunc='sum')
b =chicken.pivot_table('통화건수', index=['기준일','시군구','성별','요일','연령대'], aggfunc='sum')
c =pizza.pivot_table('통화건수', index=['기준일','시군구','성별','요일','연령대'], aggfunc='sum')
```

```python
a.reset_index(inplace=True)
a.rename(columns={'통화건수':'중식통화건수'}, inplace=True)
a.head()
```

```python
b.reset_index(inplace=True)
b.rename(columns={'통화건수':'치킨통화건수'}, inplace=True)
b.head()
```

```python
c.reset_index(inplace=True)
c.rename(columns={'통화건수':'피자통화건수'}, inplace=True)
c.head()
```

```python
d = pd.merge(a,b,how='outer',on=['기준일','시군구','성별','요일','연령대'])
e = pd.merge(d,c,how='outer',on=['기준일','시군구','성별','요일','연령대'])
e.sort_values(by=['기준일','시군구','성별','요일','연령대'], inplace=True)
e.head()
```

```python
e.isnull().sum()
```

```python
e.fillna(0, inplace=True)
e['총배달건수'] = e.eval('중식통화건수 + 치킨통화건수 + 피자통화건수')
e.loc[:,'중식통화건수':'총배달건수'] = e.loc[:,'중식통화건수':'총배달건수'].astype('int')
e.head()
```

```python
e['기준일'] = pd.to_datetime(e['기준일'].astype('str')).astype('str')
e.head()
```
